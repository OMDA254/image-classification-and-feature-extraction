```python
import os
import shutil

main_folder = "dataset2"
classes = ["cloudy", "rain", "sunrise", "shine"]

# Create class folders
for cls in classes:
    os.makedirs(os.path.join(main_folder, cls), exist_ok=True)

# Loop over files only
for filename in os.listdir(main_folder):
    file_path = os.path.join(main_folder, filename)

    # Skip if it's a directory
    if os.path.isdir(file_path):
        continue

    for cls in classes:
        if filename.lower().startswith(cls):
            dst = os.path.join(main_folder, cls, filename)
            shutil.move(file_path, dst)
            break

print("Done organizing files into class folders.")
```

```python
import os
import shutil
import random

dataset_path = "dataset2"
output_path = "weather_dataset"

classes = ["cloudy", "rain", "sunrise", "shine"]
split_ratio = 0.8  # 80% train

for cls in classes:
    class_path = os.path.join(dataset_path, cls)
    images = os.listdir(class_path)
    
    random.shuffle(images)
    
    split_index = int(len(images) * split_ratio)
    train_images = images[:split_index]
    val_images = images[split_index:]
    
    # Create folders
    os.makedirs(os.path.join(output_path, "train", cls), exist_ok=True)
    os.makedirs(os.path.join(output_path, "val", cls), exist_ok=True)
    
    # Move train images
    for img in train_images:
        src = os.path.join(class_path, img)
        dst = os.path.join(output_path, "train", cls, img)
        shutil.copy(src, dst)  
    
    # Move val images
    for img in val_images:
        src = os.path.join(class_path, img)
        dst = os.path.join(output_path, "val", cls, img)
        shutil.copy(src, dst)

print("Dataset split completed. Train and validation folders created.")
```
